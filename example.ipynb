```python
%reload_ext autoreload
%autoreload 2

import warnings
warnings.filterwarnings('ignore')

from IPython.display import Image

from app import *
import logging

logging.basicConfig(level=logging.INFO)
```

```python
index_dir = "data"
img_dirs = ["/images"]
max_images = 100
```

```python
def img_paths():
    return scan_image_dirs(img_dirs, limit=max_images)
```

```python
def display_result(query, results):
    display(query)
    for img, sim in results:
        display(sim)
        display(Image(filename=img, width=400))
```

## ClipModel

```python
app = App(ClipModel(), index_dir, batch_size=10)
```

```python
app.build_index(img_paths())
```

```python
query = "woman with child"
results = app.search(query, limit=3)
display_result(query, results)
```

```python
query = "man and woman"
results = app.search(query, limit=3)
display_result(query, results)
```

```python
query = "emotional"
results = app.search(query, limit=3)
display_result(query, results)
```
